frequency of purchaase by number of customer

```python
frequency_of_purchases_counts = df['Frequency of Purchases'].value_counts()
plt.figure(figsize=(8, 6))
frequency_of_purchases_counts.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Count of customers by gender


```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 8))
gender_counts.plot(kind='pie')
plt.title('gender')
plt.xlabel('genders')
plt.ylabel('Number of Customers')
plt.show()
```

category purchase per seasons

```python

sns.stripplot(y='Previous Purchases', x='Season', data=df, 
              jitter=True, hue='Category', dodge=True) 
```

Total Amount purchased per item

```python

plt.figure(figsize=(12, 5))
item = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()

sns.barplot(data=item,x='Item Purchased',y='Purchase Amount (USD)')
plt.title('Item purchased by Amount(USD)')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

pupular category among the customers

```python
category_counts = df['Category'].value_counts()
plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', color='lavender', edgecolor='black')
plt.title('category')
plt.xlabel('category')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

location based purchases

```python
location_counts = df['Location'].value_counts().head(10)
plt.figure(figsize=(10, 6))
location_counts.plot(kind='bar', color='purple')
plt.title('Top 10 Locations by Purchases')
plt.xlabel('Location')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

Item purchased per Season

```python
sns.countplot(y='Size', hue='Category', data=df, palette='viridis')
plt.title('Count Plot: Category vs. Season')
plt.xlabel('Count')
plt.ylabel('Category')
plt.legend(title='Season', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

highest rated purchase

```python

max_rating = df['Review Rating'].max()


highest_rated_purchase = df[df['Review Rating'] == max_rating]

ax = sns.swarmplot(x='Frequency of Purchases', y='Category', data=highest_rated_purchase) 
plt.title('Graph') 
plt.show() 
```
